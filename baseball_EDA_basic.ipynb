## 요약본

```python
def resumtable(data):
    summary = pd.DataFrame(data.dtypes).reset_index().rename(columns={"index":"피처", 0:"dtypes"})
    summary["결측치 갯수"] = data.isnull().sum().values
    summary["고윳값 갯수"] = data.nunique().values
    summary["첫번째 값"] = data.iloc[0].values
    summary["두번째 값"] = data.iloc[1].values
    summary["세번째 값"] = data.iloc[2].values
    
    print(data.shape)
    return summary
```

```python
print("salary")
resumtable(salary)
```

# 포스트시즌 데이터 만들기

```python
"""
< 포스트시즌 역사 >

1903 ~ 1968: 월드시리즈 시대
- 리그 우승팀만 포스트시즌 진출 - 2팀

1969 ~ 1993: 디비전 챔피어십 시대
- 리그를 두 개의 디비전(동부, 서부)로 나눔
- 각 리그 디비전(지역) 우승팀 포스트시즌 진출 - 4팀

1994 ~ 2011: 와일드카드 도입
- 리그를 세 개의 디비전(동부, 중부, 서부)로 재구성
- 와일드카드는 디비전 우승팀 제외 2위팀 중 가장 좋은 성적을 거둔 팀 - 8팀

2012 ~ 2021 : 현대 포스트 시즌
- 와일드카드는 디비전 우승팀 제외 가장 좋은 성적을 거둔 2팀 
- 와일드카드 팀 간 단판 승부를 통해 디비전 시리즈 진출자 결정 - 10팀(8팀)
- 포스트시즌 진출 10이 맞는데 와일드카드 2장받은사람 끼리 단판전 통해서 디비전 진출 즉 NL, AL 각각 한팀씩만 진출 (8팀)

2022 ~ 현재:
- 디비전 우승 3팀 + 와일드카드 3팀 - 12팀 진출

특징: 1904년 > 리그간 대립으로 포스트시즌 X
     1994년 > 노조 파업으로 인해 리그 리그중단, 포스트 시즌 X(~1995년)
     
< MLB 팀 개수 >
- 초창기 1903년에 16개의 팀(각 리그 8팀씩)
- 1961년을 기점으로 리그를 확장하기 시작함
- 1990년대까지 지속적인 확장으로 현재 30개 팀에 도달
- 1903년 이후 활동했던 팀(현재 + 과거)의 총수는 약 50개 이상
     
출처 : https://en.wikipedia.org/wiki/Major_League_Baseball_postseason # mlb 포스트시즌의 역사
      https://en.wikipedia.org/wiki/1904_Major_League_Baseball_season # 1904 NL 우승팀 뉴욕 자이언츠 포스트시즌 참가 거부
      https://en.wikipedia.org/wiki/1994_Major_League_Baseball_season # 1994 mlb 파업
      https://blog.naver.com/chinadrum/223575196999 # 포스트시즌 규칙
      https://ko.wikipedia.org/wiki/디비전_시리즈 # 디비전 시리즈 규칙
      https://en.wikipedia.org/wiki/Federal_League # 야구 연방 리그 (FL리그)
      https://en.wikipedia.org/wiki/1981_Major_League_Baseball_season#American_League # 1981년도 파업으로 시즌분할 > 8팀 포스트시즌 진출
"""

postseason_team = team.loc[team["연도"]>=1903] # 1903년부터 포스트시즌 존재
postseason_team = postseason_team.loc[postseason_team["연도"]!=1904] # NL 우승팀이 월드시리즈 참여 거부로 포스트시즌 열리지 않음
postseason_team = postseason_team.loc[postseason_team["연도"]!=1994] # 노조 파업으로 리그 중단, 포스트시즌 열리지 않음
postseason_team
```

```python
fig, ax = plt.subplots(2,2, figsize=(15,13.5))
# 평균 자책횟수
postseason_team.loc[(postseason_team["연도"]>=1994)].groupby(["리그", "지역"])["자책점수"].mean().plot.bar(ax=ax[0,0],
                                                                                                 color=["lightblue", "pink","lightblue", "lightblue", "lightblue"])
for idx, value in enumerate(postseason_team.loc[(postseason_team["연도"]>=1994)].groupby(["리그", "지역"])["자책점수"].mean().values):
    ax[0,0].text(
        x=idx,
        y=value+5,
        s=int(value),
        ha="center"
    )
ax[0,0].set_title("리그별 지역당 평균 자책점수")

# 평균 자책점수

postseason_team.loc[(postseason_team["연도"]>=1994)].groupby(["리그", "지역"])["평균자책점수"].mean().plot.bar(ax=ax[0,1])
for idx, value in enumerate(postseason_team.loc[(postseason_team["연도"]>=1994)].groupby(["리그", "지역"])["평균자책점수"].mean().values):
    ax[0,1].text(
        x=idx,
        y=value+0.04,
        s=f"{value:.3f}",
        ha="center"
    )
ax[0,1].set_title("리그별 지역당 평균 평균자책점수")

# 평균 실책횟수
postseason_team.loc[(postseason_team["연도"]>=1994)].groupby(["리그", "지역"])["실책횟수"].mean().plot.bar(ax=ax[1,0],
                                                                                                 color=["lightblue", "pink","lightblue", "lightblue", "lightblue"])
for idx, value in enumerate(postseason_team.loc[(postseason_team["연도"]>=1994)].groupby(["리그", "지역"])["실책횟수"].mean().values):
    ax[1,0].text(
        x=idx,
        y=value+1,
        s=f"{value:.2f}",
        ha="center"
    )
ax[1,0].set_title("리그별 지역당 평균 실책횟수")
ax[1,0].set_ylim(0,120)

# 평균 수비율

postseason_team.loc[(postseason_team["연도"]>=1994)].groupby(["리그", "지역"])["수비율"].mean().plot.bar(ax=ax[1,1])
for idx, value in enumerate(postseason_team.loc[(postseason_team["연도"]>=1994)].groupby(["리그", "지역"])["수비율"].mean().values):
    ax[1,1].text(
        x=idx,
        y=value+0.01,
        s=f"{value:.4f}",
        ha="center"
    )
ax[1,1].set_title("리그별 지역당 평균 수비율")

plt.tight_layout()
plt.show()
```

- observation
    - 우승을 많이 한 지역의 AL리그 동부지역의 평균실책횟수가 적다.
    - 해당그래프와, 바로 위 그래프(리그별 각 지역에 대한 평균 득점 관리 지표)는 팀들이 너무 많고 데이터가 충분하지 못하여 신뢰하기 어렵다는 단점이 존재

## 팀별 평균 홈런횟수

```python
plt.figure(figsize=(25,7))
data = postseason_team.groupby("팀")["홈런횟수"].mean()
lowest_team = data.nsmallest(5).index

colors = ["red" if team in lowest_team else "lightblue" for team in data.index]
data.plot.bar(color=colors)


plt.title("팀별 평균 홈런횟수(연 단위)")
plt.show()
```

- observation
    - FL리그(야구 연방 리그)의 평균자책점수와 평균 홈런횟수가 적다

```python
data = postseason_team.loc[postseason_team["연도"].isin([1914,1915])].groupby("팀")["평균자책점수"].mean()
sorted_data = data.sort_values(ascending=False)

fl_teams = postseason_team.loc[postseason_team["연도"].isin([1914,1915]) & (postseason_team["리그"]=="FL"), "팀"].unique()
colors = ["lightblue" if team in fl_teams else "pink" for team in sorted_data.index]

plt.figure(figsize=(24,6))
sorted_data.plot.bar(color=colors)

lightblue_patch = plt.Line2D([0], [0], color='lightblue', lw=4, label='FL Teams')
pink_patch = plt.Line2D([0], [0], color='pink', lw=4, label='Other Teams')
plt.legend(handles=[lightblue_patch, pink_patch], loc='upper right')
plt.title("1914~1915 팀별 평균자책점수")

plt.show()
```

## 1914~1915 팀별 평균 홈런 횟수
